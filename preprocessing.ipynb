```python
from rembg import remove
from skimage.color import rgb2gray
from skimage.filters import gaussian,laplace,threshold_otsu
from skimage.morphology import closing
from skimage.io import imshow,imread
import os
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
image = imread("me1.jpg")
imshow(image)
```

```python
image = remove(image)
imshow(image)
```

```python
from skimage.color import rgba2rgb
rgb_img = rgba2rgb(image)
rgb_img = gaussian(rgb_img)
imshow(rgb_img)
```

```python
gray_img = rgb2gray(rgb_img)
gray_img = gaussian(gray_img)
imshow(gray_img)
```

```python
bin_img = gray_img < threshold_otsu(gray_img)
imshow(bin_img)
```

```python
bin_img = closing(bin_img , np.ones((127,127)))
imshow(bin_img)
```

```python
from skimage.measure import label, regionprops, regionprops_table
```

```python
label_img = label(bin_img)
imshow(label_img)
```
